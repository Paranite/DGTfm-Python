```python
class User:
    def __init__(self, first_name, last_name, birth_year):
        self.first_name = first_name
        self.last_name = last_name
        self.birth_year = int(birth_year)
        
    def information(self):
        print("Name: {} {}, Birth year: {}".format(self.first_name, self.last_name, self.birth_year))
        
n = int(input())

userl = []
for i in range(n):
    userl.append(User(*input().split()))
    
for u in userl:
    u.information()
    
import pydoc
print(pydoc.render_doc(userl[0], renderer=pydoc.plaintext), dir(userl[0]))
```

```python
class User:
    total_users = 0
    def __init__(self, first_name, last_name, birth_year):
        self.first_name = first_name
        self.last_name = last_name
        self.birth_year = int(birth_year)
        User.total_users += 1
        
    def information(self):
        print("Name: {} {}, Birth year: {}".format(self.first_name, self.last_name, self.birth_year))
        
    @classmethod
    def from_string(cls, text):
        return cls(*text.split())
        
    def __del__(self):
        User.total_users -= 1

n = int(input())

users = {}
for i in range(n):
    cmd, uinfo = input().split(": ")
    if cmd == "Create":
        users[User.total_users] = User.from_string(uinfo)
del cmd, uinfo
    
for k, v in users.items():
    print("ID: {}, ".format(k), end = '')
    v.information()
    
print("Total users: {}".format(User.total_users))
    
import pydoc
print(pydoc.render_doc(list(users.values())[0], renderer=pydoc.plaintext), dir(list(users.values())[0]))
```

Modifikuokite 1.3 užduotyje sukurtą klasę: Pridėkite statinį metodą default_user_data, kuris visada grąžina tokį tekstą: ’User1 User2 1900’.

Modifikuokite 1.3 užduoties kodą:

Kai aptinkamas raktažodis Delete: ID, tuomet iš žodyno ištrinamas atitinkamas elementas naudojant del (7 paskaitos konspektas).
Jeigu del nesuveikia (įvyksta išimtis), tuomet except bloke nieko nedarykite (pass) ;
Kai aptinkamas raktažodis Default user, tuomet sukuriamas numatytasis naudotojas naudojant default_user_data ir from_text.

```python
class User:
    total_users = 0
    def __init__(self, first_name, last_name, birth_year):
        self.first_name = first_name
        self.last_name = last_name
        self.birth_year = int(birth_year)
        User.total_users += 1
        
    def information(self):
        print("Name: {} {}, Birth year: {}".format(self.first_name, self.last_name, self.birth_year))
        
    @classmethod
    def from_string(cls, text):
        return cls(*text.split())
    
    @staticmethod
    def default_user_data():
        return 'User1 User2 1900'
    
    def __del__(self):
        User.total_users -= 1

n = int(input())

users = {}
unum = 0
for i in range(n):
    inp = input()
    if inp == "Default user":
        unum += 1
        users[unum] = User.from_string(User.default_user_data())
        continue
    else:       
        cmd, uinfo = inp.split(": ")
        
    if cmd == "Create":
        unum +=1 
        users[unum] = User.from_string(uinfo)
    elif cmd == "Delete":
        try:
            del users[int(uinfo)]
        except:
            pass
    
for k, v in users.items():
    print("ID: {}, ".format(k), end = '')
    v.information()
    
print("Total users: {}".format(User.total_users))
    
import pydoc
print(pydoc.render_doc(list(users.values())[0], renderer=pydoc.plaintext), dir(list(users.values())[0]))
```

```python
class User:
    total_users = 0
    def __init__(self, first_name, last_name, birth_year):
        self.first_name = first_name
        self.last_name = last_name
        self.birth_year = int(birth_year)
        User.total_users += 1
        
    def information(self):
        print("Name: {} {}, Birth year: {}".format(self.first_name, self.last_name, self.birth_year))
        
    @classmethod
    def from_string(cls, text):
        return cls(*text.split())
    
    @staticmethod
    def default_user_data():
        return 'User1 User2 1900'
    
    def __del__(self):
        User.total_users -= 1
        
class Admin(User):
    def __init__(self, first_name, last_name, birth_year):
        self.privileges = ['Full']
        super().__init__(first_name, last_name, birth_year)
        
    def information(self):
        print("Name: {} {}, Birth year: {}, Privileges: {}".format(self.first_name, self.last_name, self.birth_year, self.privileges))
        

n = int(input())

users = {}
unum = 0
for i in range(n):
    inp = input()
    if inp == "Default user":
        unum += 1
        users[unum] = Admin.from_string(Admin.default_user_data()) if unum == 1 else User.from_string(User.default_user_data())
        continue
    else:       
        cmd, uinfo = inp.split(": ")
        
    if cmd == "Create":
        unum +=1 
        users[unum] = Admin.from_string(uinfo) if unum == 1 else User.from_string(uinfo)
    elif cmd == "Delete":
        try:
            del users[int(uinfo)]
        except:
            pass
    
for k, v in users.items():
    print("ID: {}, ".format(k), end = '')
    v.information()
    
print("Total users: {}".format(User.total_users))
```
